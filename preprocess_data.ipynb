```python
df['it_sim_dis_diff'] = df['it_sim_dis_diff'] - np.mean(df['it_sim_dis_diff'])
df['v2_sim_dis_diff'] = df['v2_sim_dis_diff'] - np.mean(df['v2_sim_dis_diff'])

df['it_sim_dis_diff_sq'] = [x**2 for x in df['it_sim_dis_diff']]
df['v2_sim_dis_diff_sq'] = [x**2 for x in df['v2_sim_dis_diff']]

df['it_sim_dis_diff_sq'] = df['it_sim_dis_diff_sq'] - np.mean(df['it_sim_dis_diff_sq'])
df['v2_sim_dis_diff_sq'] = df['v2_sim_dis_diff_sq'] - np.mean(df['v2_sim_dis_diff_sq'])

```

```python
df
```

```python
all_participants = set(df['participant'])

no_main_task_df = df[df['taskPhase'] == 'mainTask']

participants_with_no_main_task = set(no_main_task_df['participant'])

participants_never_main_task = all_participants - participants_with_no_main_task

print(set(participants_never_main_task))
```

```python
# print(len(set(df['participant'])))

main_task_df = df[df['taskPhase'] == 'mainTask']

participants_with_main_task_only = main_task_df.groupby('participant')['taskPhase'].apply(lambda x: (x == 'mainTask').all())

participants_with_main_task_only = participants_with_main_task_only[participants_with_main_task_only].index

print(participants_with_main_task_only)
```

```python
df = df[df['mouse.clicked_name'].apply(lambda x: len(x) > 0)]
df = df[df['taskPhase'].apply(lambda x: x == 'mainTask')]
df
```

```python
save_df = df
reliability_categorical = []
for x in save_df['reliability']:
    if x > .7:
        reliability_categorical.append(1)
    else:
        reliability_categorical.append(-1)
save_df['reliability'] = reliability_categorical

valid_map = {'valid':1,
            'invalid':-1}

save_df['validity'] = [valid_map[x] for x in save_df['validity']]

save_df['validity'] = save_df['validity'] - np.mean(save_df['validity'])
save_df['reliability'] = save_df['reliability'] - np.mean(save_df['reliability'])

save_df.to_csv('./df_pilot2_nonresp_clean.csv',
              index = False)
```
