```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

```

```python
df = pd.read_csv('new_triplets.csv')  

# Calculate correlation with p-value for statistical significance
r, p_value = stats.pearsonr(df['IT_im2_minus_im1'], df['V2_im2_minus_im1'])
print(f"Pearson r: {r:.4f}")
print(f"P-value: {p_value:.4e}")

# Create a scatter plot to visualize the relationship
plt.figure(figsize=(10, 6))
plt.scatter(df['IT_im2_minus_im1'], df['V2_im2_minus_im1'], alpha=0.6)
plt.xlabel('IT_im2_minus_im1')
plt.ylabel('V2_im2_minus_im1')
plt.title(f'IT_im2_minus_im1 vs V2_im2_minus_im1 (r = {r:.4f}, p = {p_value:.4e})')

# Add a trend line
z = np.polyfit(df['IT_im2_minus_im1'], df['V2_im2_minus_im1'], 1)
p = np.poly1d(z)
plt.plot(df['IT_im2_minus_im1'], p(df['IT_im2_minus_im1']), "r--", alpha=0.8, label='Trend line')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Optional: Calculate Spearman correlation (rank-based, robust to outliers)
spearman_r, spearman_p = stats.spearmanr(df['IT_im2_minus_im1'], df['V2_im2_minus_im1'])
print(f"\nSpearman correlation: {spearman_r:.4f}")
print(f"Spearman p-value: {spearman_p:.4e}")
```
