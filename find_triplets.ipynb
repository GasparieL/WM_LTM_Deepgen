```python
scaler = StandardScaler()

df = pd.read_csv('report.csv')
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
```

```python
df['IT_diff'] = df['abs(IT(root,img1)-IT(root,img2))']
df['V2_diff'] = df['abs(V2(root,img1)-V2(root,img2))']
```

```python
it_column = 'IT_diff'
v2_column = 'V2_diff'
img1_column = 'image1'  
img2_column = 'image2' 

it_percentile = 0.97  # Top 40%
v2_percentile = 0.9 # Bottom 60%





it_high_value = df[it_column].quantile(it_percentile)
v2_low_value = df[v2_column].quantile(v2_percentile)


if pd.api.types.is_categorical_dtype(df[it_column]) or pd.api.types.is_object_dtype(df[it_column]):
    df[it_column] = pd.to_numeric(df[it_column], errors='coerce')
if pd.api.types.is_categorical_dtype(df[v2_column]) or pd.api.types.is_object_dtype(df[v2_column]):
    df[v2_column] = pd.to_numeric(df[v2_column], errors='coerce')


it_high_value = df[it_column].quantile(it_percentile)
v2_low_value = df[v2_column].quantile(v2_percentile)
filtered_df = df[(df[it_column] >= it_high_value) & 
                 (df[v2_column] <= v2_low_value)]
print(f"Found {len(filtered_df)} rows with high IT similarity and low V2 similarity")
print(f"IT threshold (>= {it_percentile} quantile): {it_high_value}")
print(f"V2 threshold (<= {v2_percentile} quantile): {v2_low_value}")
print("\nSample of filtered data:")
# print(filtered_df)
print("\nFiltered rows with image information:")
for i, (index, row) in enumerate(filtered_df.iterrows()):
    print(f"\nRow {i+1} (Index {index}):")
    print(f"IT Similarity: {row[it_column]:.4f}, V2 Similarity: {row[v2_column]:.4f}")
    print(f"Image 1: {row[img1_column]}")
    print(f"Image 2: {row[img2_column]}")
    
    
    if i >= 9: 
        print("\n... more rows omitted ...")
        break
```
