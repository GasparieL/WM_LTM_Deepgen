```r
file_path <- "./pilot5_stats/wm_ltm_no_dist.csv"
df <- read.csv(file_path) 
# %>%
#     select(it_sim_dis_diff, it_sim_dis_diff_sq, reliability, v2_sim_dis_diff, v2_sim_dis_diff_sq, participant, validity, resp_correct, Accuracy, validity_binary, reliability_binary, v2_sim_dis_attend, v2_sim_dis_attend_z, v2_sim_dis_attend_sq_z, it_sim_dis_attend, it_sim_dis_attend_z, it_sim_dis_attend_sq_z, v2_sim_dis_unattend, v2_sim_dis_unattend_z, v2_sim_dis_unattend_sq_z, it_sim_dis_unattend, it_sim_dis_unattend_z, it_sim_dis_unattend_sq_z, it_sim_dis_diff_z, v2_sim_dis_diff_z,it_sim_dis_diff_sq_z, v2_sim_dis_diff_sq_z )
```

```r
control <- lmerControl(optimizer = "bobyqa")#, optCtrl = list(maxfun = 10000))
```

```r
formula1 <- glmer(
  recall_answ ~ resp_correct + 
    mem_stim_type * validity_binary * reliability_binary * it_sim_dis_diff_z * v2_sim_dis_diff_z
    + (1 | participant),
  data = df,
  family = binomial, control = glmerControl(optimizer = "bobyqa"))


# View the model summary
summary(formula1)

### here recall_answ = 1 if ltm task correct, 0 if its wrong
### resp_correct = 1 if wm correct
### category : tested, not tested, distractor, foil?
### and all columns are sometimes NaN - like validity for foils or distractors, etc
```

```r
tidy_model <- broom::tidy(formula1)
significant_terms <- tidy_model %>% filter(p.value < 0.05)
print(significant_terms)
```

```r


# Assuming your model is called `model`
tidy_results <- tidy(formula1, effects = "fixed", conf.int = TRUE)

# Filter for significant terms (e.g., p < 0.05)
significant_results <- tidy_results %>% filter(p.value < 0.05)

print(significant_results)
```

```r

stargazer(formula1, type="text", digits=2)
```

```r


# Create a compact summary table
tab_model(formula1,
          show.ci = FALSE,          # Hide confidence intervals if not needed
          show.se = TRUE,           # Show standard errors
          p.style = "stars",        # Add stars for significance levels
          digits = 3)               # Round values to 3 decimal places
```

```r
# Get the summary of the model
model_summary <- summary(formula1)

# Extract fixed effects
fixed_effects <- as.data.frame(model_summary$coefficients)

# Filter for significant terms (p < 0.05)
significant_terms <- fixed_effects[fixed_effects[, "Pr(>|z|)"] < 0.05, ]

# Print the filtered results
print(significant_terms)
```
