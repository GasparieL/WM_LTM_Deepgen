```python
import os
import random
import pandas as pd
import shutil
```

```python
# Define the folder path and CSV file name
folder_path = './stimuli'
folder_path_foils = './stimuli/new_foil_stims'
output_csv = 'selected_and_foils.csv'
output_folder = 'selected_and_foils'

if os.path.exists(output_folder):
    shutil.rmtree(output_folder)  # This will delete the entire folder and its contents

# Create the output folder after removing the old one
os.makedirs(output_folder)

# Remove the output CSV if it exists
if os.path.exists(output_csv):
    os.remove(output_csv)
    
# Get all files in the folder
all_images = [os.path.join(folder_path, img) for img in os.listdir(folder_path) if img.endswith(('.jpg'))]
all_images_foils = [os.path.join(folder_path_foils, img) for img in os.listdir(folder_path_foils) if img.endswith(('.jpg'))]

# Select 120 random images
selected_images = random.sample(all_images, 120)
selected_images_foils = random.sample(all_images_foils, 120)
# Create a DataFrame with a single column called ImagePath
combined_images = selected_images + selected_images_foils
random.shuffle(combined_images)

# Create a DataFrame with a single column called ImagePath
df = pd.DataFrame(combined_images, columns=['ImagePath'])

df.to_csv(output_csv, index=False)

# print(f'Selected 240 random images and saved to {output_csv}')
```

```python
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
    
new_image_paths = []
for img in combined_images:
    new_img_path = os.path.join(output_folder, os.path.basename(img))
    shutil.copy(img, new_img_path)  # Copy the image to the new folder
    new_image_paths.append(new_img_path)
```
