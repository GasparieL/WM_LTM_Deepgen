# Visualizing IT LTM Difference vs P-Values

This notebook visualizes the relationship between `it_ltm_diff` and flipped p-values.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Set style
sns.set_style('whitegrid')
sns.set_context('notebook', font_scale=1.2)
```

## Load Data

```python
# Load the two dataframes
df1 = pd.read_csv('pilot_total_matched_closest.csv')
df2 = pd.read_csv('pilot_total_ltm_stats/TEST_wm_ltm_p_pilot_total_with_predictions.csv')

print(f"Dataset 1 shape: {df1.shape}")
print(f"Dataset 2 shape: {df2.shape}")
```

## Plot 1: it_ltm_diff vs p_values_it_flipped_ltm

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.pointplot(
    data=df1,
    x='it_ltm_diff',
    y='p_values_it_flipped_ltm',
    ax=ax,
    estimator=np.mean,
    errorbar='se',
    capsize=0.1,
    color='steelblue'
)

# Customize plot
ax.set_xlabel('IT LTM Difference', fontsize=14)
ax.set_ylabel('P-values IT Flipped LTM', fontsize=14)
ax.set_title('Closest Point Matching: IT LTM Difference vs P-values IT Flipped LTM', fontsize=16, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
ax1.plot(df2['it_ltm_diff'], df2['p_flipped_it_test_new'], 'o', 
         color='#49a791', alpha=0.6, markersize=6)
ax1.axvline(0, color='gray', linestyle='--', alpha=0.5)
ax1.set_xlabel('IT Distractor Similarity Difference (Test)')
ax1.set_ylabel('Predicted Accuracy (p_it_test_new)')
ax1.set_title('Model point matching: IT Model Predictions')
ax1.grid(True, alpha=0.3)

ax2.plot(df2['v2_ltm_diff'], df2['p_flipped_v2_test_new'], 'o', 
         color='#95c355', alpha=0.6, markersize=6)
ax2.axvline(0, color='gray', linestyle='--', alpha=0.5)
ax2.set_xlabel('V2 Distractor Similarity Difference (Test)')
ax2.set_ylabel('Predicted Accuracy (p_v2_test_new)')
ax2.set_title('Model point matching: V2 Model Predictions')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
ax1.plot(df1['it_ltm_diff'], df1['p_values_it_flipped_ltm'], 'o', 
         color='#49a791', alpha=0.6, markersize=6)
ax1.axvline(0, color='gray', linestyle='--', alpha=0.5)
ax1.set_xlabel(' IT Distractor Similarity Difference (Test)')
ax1.set_ylabel('Predicted Accuracy (p_it_test_new)')
ax1.set_title('Closest Point Matching: IT Model Predictions')
ax1.grid(True, alpha=0.3)

ax2.plot(df1['v2_ltm_diff'], df1['p_values_v2_flipped_ltm'], 'o', 
         color='#95c355', alpha=0.6, markersize=6)
ax2.axvline(0, color='gray', linestyle='--', alpha=0.5)
ax2.set_xlabel('V2 Distractor Similarity Difference (Test)')
ax2.set_ylabel('Predicted Accuracy (p_v2_test_new)')
ax2.set_title('Closest Point Matching: V2 Model Predictions')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Plot 2: it_ltm_diff vs p_flipped_it_test_new

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.pointplot(
    data=df2,
    x='it_ltm_diff',
    y='p_flipped_it_test_new',
    ax=ax,
    estimator=np.mean,
    errorbar='se',
    capsize=0.1,
    color='coral'
)

ax.set_xlabel('IT LTM Difference', fontsize=14)
ax.set_ylabel('P Flipped IT Test New', fontsize=14)
ax.set_title('Model point matching: IT LTM Difference vs P Flipped IT Test New', fontsize=16, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Combined Plot (Optional)

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Plot 1
sns.pointplot(
    data=df1,
    x='it_ltm_diff',
    y='p_values_it_flipped_ltm',
    ax=axes[0],
    estimator=np.mean,
    errorbar='se',
    capsize=0.1,
    color='steelblue'
)
axes[0].set_xlabel('IT LTM Difference', fontsize=12)
axes[0].set_ylabel('P-values IT Flipped LTM', fontsize=12)
axes[0].set_title('Closest Point matching: Dataset 1: it_ltm_diff vs p_values_it_flipped_ltm', fontsize=13, fontweight='bold')
axes[0].tick_params(axis='x', rotation=45)

# Plot 2
sns.pointplot(
    data=df2,
    x='it_ltm_diff',
    y='p_flipped_it_test_new',
    ax=axes[1],
    estimator=np.mean,
    errorbar='se',
    capsize=0.1,
    color='coral'
)
axes[1].set_xlabel('IT LTM Difference', fontsize=12)
axes[1].set_ylabel('P Flipped IT Test New', fontsize=12)
axes[1].set_title('Model point matching: Dataset 2: it_ltm_diff vs p_flipped_it_test_new', fontsize=13, fontweight='bold')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
