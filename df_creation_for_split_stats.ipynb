```python
# Import required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os.path as op
import glob
from datetime import timedelta
from sklearn.preprocessing import StandardScaler
import os
```

```python
df_total = pd.read_csv('pilot_total.csv')
# (pilot5.csv, pilot6.csv)

df_correct = df_total[df_total['resp_correct'] == 1]
df_correct['rt_log'] = np.log(df_correct['rt'])
df_incorrect =  pd.read_csv('pilot_total_errors.csv')

df_names = {
    id(df_total): 'total',
    id(df_correct): 'correct', 
    id(df_incorrect): 'incorrect'
}
```

```python
for df in [df_total, df_correct, df_incorrect]:
    
    df_it_abs_neg = df[df['it_sim_dis_test'] < 0]
    df_it_abs_pos = df[df['it_sim_dis_test'] > 0]
    df_v2_abs_neg = df[df['v2_sim_dis_test'] < 0]
    df_v2_abs_pos = df[df['v2_sim_dis_test'] > 0]
    
    df_it_rel_neg = df[df['it_sim_dis_diff_test'] < 0]
    df_it_rel_pos = df[df['it_sim_dis_diff_test'] > 0]
    df_v2_rel_neg = df[df['v2_sim_dis_diff_test'] < 0]
    df_v2_rel_pos = df[df['v2_sim_dis_diff_test'] > 0]
   
    def flip_z_sq_z_cube(df, column_name):
        scaler_reliability = StandardScaler()
        df['reliability_binary_z'] = scaler_reliability.fit_transform(
            pd.DataFrame(df['reliability_binary'])
        ).flatten()

        scaler_validity = StandardScaler()
        df['validity_binary_z']  = scaler_validity.fit_transform(
            pd.DataFrame(df['validity_binary'])
        ).flatten()

        
        df_flip = df.copy()
        new_columns = {}
        new_columns[column_name] = df[column_name]
    
        scaler_z = StandardScaler()
    
        new_columns[column_name + '_z'] = scaler_z.fit_transform(
            pd.DataFrame(new_columns[column_name])
        ).flatten()
        scaler_sq_z = StandardScaler()
    
        new_columns[column_name + '_sq'] = new_columns[column_name + '_z'] ** 2
        new_columns[column_name + '_sq_z'] = scaler_sq_z.fit_transform(
            pd.DataFrame(new_columns[column_name + '_sq'])
        ).flatten()
        
        scaler_cube_z = StandardScaler()
        new_columns[column_name + '_cube'] = new_columns[column_name + '_z'] ** 3
        new_columns[column_name + '_cube_z'] = scaler_cube_z.fit_transform(
            pd.DataFrame(new_columns[column_name + '_cube'])
        ).flatten()
        
        for col_name, values in new_columns.items():
            df_flip[col_name] = values
        
        return df_flip
    
    for column in ['it_sim_dis_diff_test', 'v2_sim_dis_diff_test', 'it_sim_dis_test', 'v2_sim_dis_test']:
        df_it_rel_neg = flip_z_sq_z_cube(df_it_rel_neg, column)
        df_it_rel_pos = flip_z_sq_z_cube(df_it_rel_pos, column)
        df_v2_rel_neg = flip_z_sq_z_cube(df_v2_rel_neg, column)
        df_v2_rel_pos = flip_z_sq_z_cube(df_v2_rel_pos, column)

    df_name = df_names[id(df)]
    
    # Create directory if it doesn't exist
    os.makedirs(f'stats_wm/pilot6/{df_name}', exist_ok=True)
    
    # Save the correct dataframes to CSV files
    df_it_rel_neg.to_csv(f'stats_wm/total/{df_name}/df_it_rel_neg.csv', index=False)
    df_it_rel_pos.to_csv(f'stats_wm/total/{df_name}/df_it_rel_pos.csv', index=False)
    df_v2_rel_neg.to_csv(f'stats_wm/total/{df_name}/df_v2_rel_neg.csv', index=False)
    df_v2_rel_pos.to_csv(f'stats_wm/total/{df_name}/df_v2_rel_pos.csv', index=False)
```
