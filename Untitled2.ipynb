```python

df = pd.read_csv('pilot_total.csv')
df_quest = pd.read_csv("questionnaires.csv")

total_participants = set(df['participant'])
df_new_data = df[df['pilot_number'] == 1]
df_old_data = df[df['pilot_number'] == 2]
new_data_participants = set(df_new_data['participant'])
old_data_participants = set(df_old_data['participant'])
```

```python
df_quest_new = df_quest[df_quest['id'].isin(new_data_participants)]
df_quest_old = df_quest[df_quest['id'].isin(old_data_participants)]
```

```python
df_quest_clean = df_quest[~df_quest['id'].isin(['id', '{"ImportId":"id"}'])]
df_quest_clean = df_quest_clean[df_quest_clean['id'].str.isnumeric()]

df_quest_clean['id'] = df_quest_clean['id'].astype(int)

print(f"Cleaned df_quest size: {len(df_quest_clean)}")
print(f"Sample cleaned IDs: {list(df_quest_clean['id'].head(10))}")

overlap = total_participants.intersection(set(df_quest_clean['id']))
print(f"Overlapping IDs after cleaning: {len(overlap)}")

df_quest_new = df_quest_clean[df_quest_clean['id'].isin(new_data_participants)]
df_quest_old = df_quest_clean[df_quest_clean['id'].isin(old_data_participants)]

print(f"New data questionnaires: {len(df_quest_new)}")
print(f"Old data questionnaires: {len(df_quest_old)}")
```

```python
df_quest_clean = df_quest_clean.drop_duplicates(subset='id', keep='first')

print(f"After removing duplicates: {len(df_quest_clean)}")
print(f"Unique IDs: {df_quest_clean['id'].nunique()}")

df_quest_new = df_quest_clean[df_quest_clean['id'].isin(new_data_participants)]
df_quest_old = df_quest_clean[df_quest_clean['id'].isin(old_data_participants)]

print(f"New data questionnaires: {len(df_quest_new)}")
print(f"Old data questionnaires: {len(df_quest_old)}")

print(f"Unique participants in df_quest_new: {df_quest_new['id'].nunique()}")
print(f"Unique participants in df_quest_old: {df_quest_old['id'].nunique()}")
```
