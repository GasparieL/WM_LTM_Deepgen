```python
import pandas as pd
 
```

```python
# Load the old CSV file named 'condition'
file_path = 'conditions_file_new_stims.csv'

# Read the CSV file
condition_df = pd.read_csv(file_path)

# Display the content of the CSV file to understand its structure
condition_df.head()
```

```python

condition_df['Image1'] = condition_df['img1'].apply(lambda x: x.split('/')[2] if isinstance(x, str) and len(x.split('/')) > 2 else '')

# Image2 should be everything after the second '/' in the 'img2' column
condition_df['Image2'] = condition_df['img2'].apply(lambda x: x.split('/')[2] if isinstance(x, str) and len(x.split('/')) > 2 else '')
condition_df['IT_most_similar'] = condition_df.apply(
    lambda row: 'im1' if row['IT_root_im1'] > row['IT_root_im2'] else 'im2', axis=1
)
condition_df['V2_most_similar'] = condition_df.apply(
    lambda row: 'im1' if row['V2_root_im1'] > row['V2_root_im2'] else 'im2', axis=1
)


condition_df['IT_im2_minus_im1'] = condition_df['IT_root_im2'] - condition_df['IT_root_im1']
condition_df['V2_im2_minus_im1'] = condition_df['V2_root_im2'] - condition_df['V2_root_im1']


condition_df.to_csv('conditions_file_new_stims.csv', index=False)


# Provide the user with the updated file
condition_df
```

```python


#Create a new DataFrame using the same data from the old file but with the new column names provided earlier
new_columns = [
    'IT_diff', 'V2_diff', 'root', 'Image1', 'Image2', 'IT_root_im1', 'IT_root_im2',
    'IT_im1_im2', 'V2_root_im1', 'V2_root_im2', 'V2_im1_im2', 'cat_root', 'cat_im1',
    'cat_im2', 'V2_im2_minus_im1',
    'IT_im2_minus_im1', 'IT_most_similar', 'V2_most_similar'
]

# Filter or rearrange the columns of the old dataframe to match the new column structure
triplets_df = condition_df[new_columns]

### I dont have in new conditions file a concept_root, concept_im1 and concept_im2 also no V2_im2_minus_im1 and IT_im2_minus_im1 
### also no IT_most_similar	V2_most_similar but I will be adding those now and that just is wrt root
### also adding IT_im2_minus_im1 and V2_im2_minus_im1
### triplets only missing concept columns which just are a general category description
```

```python

triplets_df['root'] = triplets_df['root'].apply(lambda x: x.split('/')[2] if isinstance(x, str) and len(x.split('/')) > 2 else '')
```

```python

triplets_df['root_numeric'] = triplets_df['root'].apply(lambda x: int(x.split('_')[0]) if isinstance(x, str) else 0)

# Sort the dataframe by the 'root_numeric' column
triplets_df = triplets_df.sort_values(by='root_numeric').drop(columns=['root_numeric'])
```

```python
triplets_df.to_csv('new_triplets.csv', index=False)

```
